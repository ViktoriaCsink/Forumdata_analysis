```python
def clean_text(content):
    
    if type(content) == bytes:
        content = content.decode('utf-8')
    content = re.findall(r'[a-zA-Z]+', content)
    content = [c.lower() for c in content]
    
    #rm non-English words and stop-words
    en_words = set(nltk.corpus.words.words())
    stop_words = [s for s in stopwords.words('english')]
    content = [c for c in content if c in en_words and c not in stop_words]
    
    #rm very short words
    content = [c for c in content if len(c)>=3]
    
    #join it back into a tidy string of words
    content = ' '.join([c for c in content])
    
    return content
```

```python
forumdata['cleaned_message'] = [clean_text(m) for m in tqdm(forumdata.raw_message)]
```

### Create a dictionary with the afinn scores

```python
os.chdir(base)
with open('afinn-165.txt', 'r') as f:
    wordlist = f.readlines()
    afinn = {}
```

```python
for w in wordlist:
    try:
        word = re.findall('^(.*?):', w)[0].strip() #get the word
        num = re.findall(':(.*)', w)[0].strip() #get the score
        num = re.sub(r',', '', num)
        afinn[word] = int(num)
    except:
        wordlist.remove(w)
```

### Add up the raw afinn scores of each word

```python
def count_raw_scores(message):
    
    scores = []
    words = [w.lower() for w in message.split()]
    
    for w in words:
        if w in afinn.keys():
            scores.append(afinn.get(w))
        else:
            scores.append(0)
    
    if scores:
        total_score = sum(scores)
    else:
        total_score = None
    
    return scores, total_score
```

```python
forumdata['raw_afinn_scores'] = ''
for r,v in tqdm(forumdata.iterrows()):
    
    scores, total_score = count_raw_scores(forumdata.loc[r, 'cleaned_message'])
    forumdata['raw_afinn_scores'][r] = scores
    forumdata.loc[r, 'total_raw_afinn_score'] = total_score
    
forumdata = forumdata.dropna()
forumdata = forumdata.reset_index(drop=True)
forumdata.head()
```

### Get the tfidf scores of all the words in the message, as well as the afinn scores of the same words. Take the dot product to express both how rare a word is as well as the sentiment. Words like 'good' will be weeded out.

```python
cleaned_data = []
cleaned_data.extend(''.join(forumdata['cleaned_message'][index]) for index, row in forumdata.iterrows())

tf = TfidfVectorizer()

cleaned_data_transformed = tf.fit_transform(cleaned_data)

data_vector = pd.DataFrame(cleaned_data_transformed.toarray(),
                 columns=tf.get_feature_names())

data_vector.shape
```

```python
data_vector.head()
```

### Get the tfidf scores of each word in the order of ocurrence in the message

```python
forumdata['tfidf_scores'] = ''
for r,v in tqdm(forumdata.iterrows()):
    tfidfs = []
    
    #Get the tfidf score for each word in the message
    for word in forumdata.loc[r, 'cleaned_message'].split():
        tfidfs.append(data_vector.loc[r, :].get(word))
        forumdata['tfidf_scores'][r] = tfidfs
```

### Take the product of the tfidf values and the afinn scores for each word and add this onto forumdata

```python
forumdata['corrected_afinn_scores'] = ''
for r,v in tqdm(forumdata.iterrows()):
    
    raw=forumdata.loc[r, 'raw_afinn_scores']
    tfidfs = forumdata.loc[r, 'tfidf_scores']
    forumdata['corrected_afinn_scores'][r] = np.asarray(raw)*np.asarray(tfidfs)
```

### Finally, calculate the sum of the corrected afinn scores to represent the valence of the entire message

```python
forumdata['total_corrected_afinn_score'] = forumdata.apply(lambda x: np.sum(x.corrected_afinn_scores), axis=1)
forumdata = forumdata.sort_values(by=['total_corrected_afinn_score'], ascending=True)
```

Extract the negative and positive words from each message

```python
sentwords = pd.DataFrame()
row=0

for r,v in tqdm(top_forumdata.iterrows()):
    
    message_words = top_forumdata.loc[r, 'cleaned_message'].split()

    #get unique words, record the frequency and the name of the forum
    for word in set(message_words):
        
        if word in afinn.keys():
        
            sentwords.loc[row, 'forum'] = top_forumdata.loc[r, 'topic_url'] #topic
            sentwords.loc[row, 'word'] = word
            sentwords.loc[row, 'freq'] = collections.Counter(message_words).get(word) #frequency in post
            row = row+1
            
sentwords = pd.DataFrame(sentwords.groupby(['forum', 'word']).freq.sum()).reset_index(drop=False)
sentwords['raw_score'] = sentwords['word'].apply(lambda x: afinn.get(x)) #afinn score
sentwords = sentwords.sort_values(by=['forum']).reset_index(drop=True)
```

```python
#Calculate a unique colour for each afinn score to express degree of negative/positive

def get_opacity(score):

    #abs value of all sentiment scores
    degrees = [abs(v) for k,v in afinn.items() if v!=0]

    gradient = 1/max(degrees)
    
    #1 if most neg/pos, reduced opacity otherwise
    alpha = abs(score)*gradient

    return alpha


def get_colour(word):
    
    score = afinn.get(word)
    alpha = get_opacity(score)
    
    if score > 0: #blue if pos
        colour = (0, 0, 1, alpha) 
    elif score < 0: #red if neg
        colour = (1, 0, 0, alpha)
        
    return colour
```

```python
sentwords['colour'] = sentwords['word'].apply(get_colour)
```

### 2. Once sentiment words are removed, is there a difference between the frequency of the non-sentiment words in the positive and negative messages?

```python
top_forumdata.topic_url.unique().tolist()
```

```python
neg_data = pd.DataFrame(columns=top_forumdata.columns)
pos_data = pd.DataFrame(columns=top_forumdata.columns)

for f in top_forumdata.topic_url.unique():
    
    this_forum = top_forumdata[top_forumdata.topic_url==f]
    M = mean(this_forum.total_corrected_afinn_score)
    sd = stdev(this_forum.total_corrected_afinn_score)
    
    neg_data = neg_data.append(this_forum[this_forum.total_corrected_afinn_score<(M-sd)])
    pos_data = pos_data.append(this_forum[this_forum.total_corrected_afinn_score>(M+sd)])
```

```python
#Words in a positive context with frequency count
pos_words = dict(collections.Counter(pos_text.split()))
pos_words = pd.DataFrame.from_dict(pos_words, orient='index', columns = ['pos_count'])

#Words in a negative context with frequency count
neg_words = dict(collections.Counter(neg_text.split()))
neg_words = pd.DataFrame.from_dict(neg_words, orient='index', columns = ['neg_count'])

word_count = pos_words.join(neg_words)
word_count = word_count.rename_axis('word').reset_index()
word_count = word_count.dropna()

#order them by difference and show the top 50
word_count['diff'] = word_count.apply(lambda x: abs(x.pos_count - x.neg_count), axis=1)
word_count = word_count.sort_values(by = ['diff'], ascending=False).reset_index(drop=True)

os.chdir(nonsents)
word_count.to_csv('Freq_non_sentiment_words.csv')
os.chdir(preproc)
word_count.to_csv('Freq_non_sentiment_words.csv')

#Take the top 50 words where differences are largest (for plotting)
word_count = word_count[0:50]

word_count.head()
```

Plot the words where the context mattered the most.

```python
os.chdir(nonsents)

#words
labels = [word for word in word_count.word]

#array where the first value is pos, second neg (neg values are subtracted from 0 to appear on the left side of the x axis)
data = np.array([[a, 0-b] for a,b in zip(word_count.pos_count.tolist(), word_count.neg_count.tolist())])

#the borders of the graph
data_cum = data

category_names = ['Positive context', 'Negative context']

category_colors = [(0, 0.28, 0.73, 1), (1, 0, 0, 1)]

fig, ax = plt.subplots(figsize=(16, 40))

ax.invert_yaxis()
ax.xaxis.set_visible(False)

#set the same maximum distance from 0
if abs(min(data[:,1])) >= max(data[:,0]):
    ax.set_xlim(min(data[:,1]), abs(min(data[:,1])))
else:
    ax.set_xlim(0-max(data[:,0]), max(data[:,0]))
       

ax.set_ylim(-1,50)
ax.set_yticklabels(labels, fontsize=14)

#go left and right from 0
for i, (colname, color) in enumerate(zip(category_names, category_colors)):
    widths = data[:, i]
    starts = data_cum[:, i] - widths
    rects = ax.barh(labels, widths, left=starts, height=0.5,
                    label=colname, color=color)
    
    r, g, b, _ = color
    text_color = 'white' if b > 0.5 else 'black' #see what's most visible
    
    #display the absolute values, as frequency count can't be negative
    ax.bar_label(rects, labels = abs(data[:,i]), label_type='center', color=text_color)
        
ax.legend(ncol=len(category_names), bbox_to_anchor=(0,1), loc='lower left', fontsize='large')

plt.title('"Where context matters the most"\n\nLargest differences in frequency between positive and negative messages\nAIBU (169 + posts, 160 - posts)\n\n\n', fontsize=15)

plt.savefig('Differences_non_sentwords.pdf')
plt.show()

    
```
